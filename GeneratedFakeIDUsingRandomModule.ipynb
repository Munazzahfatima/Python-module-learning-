# ***USING RANDOM MODULE***
# ***TO GENERATING A FAKE ID***

```python
import random
first_names = ["Ayesha", "Ahmed", "Fatima", "Hassan", "Zainab", "Ali", "Maryam", "Bilal", "Sara", "Usman"]

last_names = ["Khan", "Ali", "Shah", "Iqbal", "Malik", "Raza", "Siddiqui", "Chaudhry", "Javed", "Ahmed"]

street_names = ["Main Boulevard","Iqbal Avenue","Jinnah Road","Shahrah-e-Faisal","Mall Road","Firdous Street","Gulberg Lane","Allama Iqbal Road","Garden Street","Liberty Avenue","Model Town Boulevard","Canal View Road","Rehman Street","Shaheen Avenue","Park Lane"]

city_names = ["Karachi","Lahore","Peshawar","Quetta","Faisalabad","Rawalpindi","Multan","Sialkot","Hyderabad","Islamabad","Gujranwala","Bahawalpur","Mardan","Abbottabad"]

for num in range (50):      #generate 50 fake profiles
    first = random.choice(first_names)
    last =random.choice(last_names)
    phone=f'{random.randint(100,999)}-111-{random.randint(1000,9999)}'
    street_num=random.randint(10,100)
    street=random.choice(street_names)
    city=random.choice(city_names)
    zip_code=random.randrange(100,1000)
    house_no=random.randrange(1,1000)

    address=f'city : {city}, zip code: {zip_code}, street: {street_num} {street} ,house no: {house_no}'
    email=first.lower() +last.lower() + '@gmail.com'

    print(f'-> {first} {last}\n {address} \n{email}\n')


```
